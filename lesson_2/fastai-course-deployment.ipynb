```python
# Download and decompress our dataset
path = untar_data(URLs.PETS)/'images'
```

```python
# to label check if filename starts with a capital letter
def is_cat(x): return x[0].isupper()
```

```python
# train with resnet18
data_loaders = ImageDataLoaders.from_name_func('.', get_image_files(path), valid_pct=0.2, seed=42, label_func=is_cat, item_tfms=Resize(192))
learn = vision_learner(data_loaders, resnet18, metrics=error_rate)
learn.fine_tune(3)
```

```python
# export the model
learn.export("model.pkl")

# use the model
im = PILImage.create("dog.jpg")
im.thumbnail((192, 192))

learn = load_learner("model.pkl")
learn.predict(im)
```
